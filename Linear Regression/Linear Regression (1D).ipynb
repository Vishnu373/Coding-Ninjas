```python
import numpy as np
```

```python
# load the data
data = np.loadtxt("https://files.codingninjas.in/data-6984.csv", delimiter = ",")
```

```python
# split it into input and output or x and y
x = data[:, 0].reshape(-1, 1)
y = data[:, 1]
```

```python
# splitting x into training and testing data
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x, y)
```

```python
# importing linear regression
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
```

```python
# training the model with the trainind data
"""
alg1.fit(X_train, Y_train)
But this will throw an error, as it requires 2D array or more.
Hence, perform reshape(-1, 1)
"""

# after reshaping
alg1.fit(X_train, Y_train)
```

```python
# checking m and c i.e., coefficient and intercept
print(alg1.coef_)
print(alg1.intercept_)
```

```python
# plotting the graph with m and c values
import matplotlib.pyplot as plt

m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(30, 70, 0.1)
y_line = m * x_line + c

plt.plot(x_line, y_line)

train_1d = X_train.reshape(75)
plt.scatter(X_train, Y_train)

plt.show()
```

```python
# plotting the graph for testing data
import matplotlib.pyplot as plt

m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(30, 70, 0.1)
y_line = m * x_line + c

plt.plot(x_line, y_line)

train_1d = X_test.reshape(25)
plt.scatter(X_test, Y_test)

plt.show()
```

```python
# finding score or coefficient of determination
score_training = alg1.score(X_train, Y_train)
score_test = alg1.score(X_test, Y_test)

print(score_training)
print(score_test)
```
