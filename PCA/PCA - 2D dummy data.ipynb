```python
import numpy as np
import matplotlib.pyplot as plt

# PCA import
from sklearn.decomposition import PCA
```

```python
x1 = np.array([1, 2, 3, 4, 5, 6])
x2 = np.array([7.5, 11, 16, 18, 20, 26])
```

```python
plt.scatter(x1, x2)
plt.show()
```

```python
# combining both arrays to form a 2D array
X = np.c_[x1, x2]
```

```python
# performing pca
# pca = PCA()
pca = PCA(n_components = 1) # 1D case, reducing the dimensionality. Here, it is 2D to 1D

# pca = PCA(n_components = 2) # temporary case

X_transformed = pca.fit_transform(X)

X_transformed

# fit_transform() -> gives the transformed version of the given data
```

```python
# checking directions of our data
pca.components_
```

```python
# reversing the convertion, temporary case and 1D case
x_approx = pca.inverse_transform(X_transformed)
x_approx
```

```python
plt.scatter(x_approx[:, 0], x_approx[:, 1])
plt.show()
```
