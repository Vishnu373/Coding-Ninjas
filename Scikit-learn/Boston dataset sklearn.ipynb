```python
# Alternate method for loading boston data
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
X = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
Y = raw_df.values[1::2, 2]
```

```python
X.shape
```

```python
Y.shape
```

```python
# converting the data into data frame
df = pd.DataFrame(X)
df
```

```python
# changing the column names
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
df.describe()
```

```python
# training the dataset
# 1. spliting the data for training and testing
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y)
```

```python
# checking the sizes
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# 2. importing Linear Regression algorithm
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
```

```python
# 3. training the model with the training data and linear regression algorithm
alg1.fit(X_train, Y_train)
```

```python
# 4. now testing the model with the testing data and linear regression algorithm
Y_pred = alg1.predict(X_test)
```

```python
# 5. compare the predicted data and output of the testing data
# for this, plot the graph

import matplotlib.pyplot as plt

plt.scatter(Y_test, Y_pred)
plt.axis([0, 40, 0, 40])
plt.show()
```
