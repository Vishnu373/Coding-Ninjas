```python
from sklearn import datasets
```

```python
# loading the file
cancer_ds = datasets.load_breast_cancer()
```

```python
# splitting into input and output
X = cancer_ds.data
Y = cancer_ds.target
```

```python
X.shape, Y.shape
```

```python
# converting X into datdframes
import pandas as pd
cancer_df = pd.DataFrame(X)
```

```python
cancer_df
```

```python
cancer_ds.feature_names
```

```python
# setting column names
cancer_df.columns = cancer_ds.feature_names
```

```python
cancer_df
```

```python
# splitting X and Y into training and testing
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y)
```

```python
X_train.shape, Y_train.shape
```

```python
# importing the algorithm
from sklearn.linear_model import LogisticRegression
alg1 = LogisticRegression()
```

```python
# training the model
alg1.fit(X_train, Y_train)
```

```python
# Testing the model using test data
alg1.predict(X_test)
```

```python
print("Score on training data: ", alg1.score(X_train, Y_train))
print("Score on training data: ", alg1.score(X_test, Y_test))
```

```python
# chedcking the error count
alg1.predict(X_test) - Y_test
```
