```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
# laoding training data
titanic_trd = pd.read_csv("titanic_train.csv") 
```

```python
titanic_trd
titanic_trd.describe()
```

```python
# deleting columns
titanic_trd.drop(["Name", "Ticket"], axis = 1, inplace = True)
```

```python
def male_num(s):
    if s == "male":
        return 1
    else:
        return 0

def female_num(s):
    if s == "female":
        return 1
    else:
        return 0
```

```python
# converting columns
titanic_trd["Male"] = titanic_trd["Sex"].apply(male_num)
titanic_trd["Female"] = titanic_trd["Sex"].apply(female_num)
```

```python
titanic_trd.drop("Sex", axis = 1, inplace = True)
```

```python
def cherbourg(s):
    if s=="C":
        return 1
    else:
        return 0
    
def queenstown(s):
    if s=="Q":
        return 1
    else:
        return 0

def southamptom(s):
    if s=="S":
        return 1
    else:
        return 0
```

```python
titanic_trd["Cherbourg"] = titanic_trd.Embarked.apply(cherbourg)
titanic_trd["Queenstown"] = titanic_trd.Embarked.apply(queenstown)
titanic_trd["Southamptom"] = titanic_trd.Embarked.apply(southamptom)
```

```python
titanic_trd.drop("Embarked", axis = 1, inplace = True)
```

```python
titanic_trd
```

```python
def cabin_num(x):
    if x is np.nan:
        return 0
    else:
        return 1
```

```python
titanic_trd["Cabin"] = titanic_trd.Cabin.apply(cabin_num)
```

```python
titanic_trd["Age"].fillna(titanic_trd["Age"].mean(), inplace = True)
```

```python
titanic_trd
```

```python
# re arranging the order of features
titanic_trd = titanic_trd[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare','Cabin','Male','Female','Cherbourg','Queenstown','Southamptom','Survived']]
```

```python
titanic_trd
```

```python
# to numpy arrays
titanic_arr = titanic_trd.values
```

```python
titanic_arr.shape
```

```python
# splitting into input and output
X = titanic_arr[:,0:11]
Y = titanic_arr[:,11]
X.shape,Y.shape
```

```python
scaler = preprocessing.StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
X_scaled.shape
```

```python
X_scaled = np.append(X_scaled, np.ones(X_scaled.shape[0]).reshape(-1,1), axis=1)
```

```python
X_scaled.shape,X_scaled[0].shape
```

```python
# importing the algorithm
from sklearn.linear_model import LogisticRegression
alg = LogisticRegression()
```

```python
# training the model
alg.fit(X_scaled, Y)
```

```python
# testing dataset
titanic_ted = pd.read_csv("titanic_test.csv")
```

```python
titanic_ted.drop(["Name","Ticket"], axis = 1, inplace = True)
```

```python
titanic_ted["Male"] = titanic_ted["Sex"].apply(male_num)
titanic_ted["Female"] = titanic_ted["Sex"].apply(female_num)
```

```python
titanic_ted.drop("Sex",axis=1,inplace=True)
```

```python
titanic_ted["Cherbourg"] = titanic_ted.Embarked.apply(cherbourg)
titanic_ted["Queenstown"] = titanic_ted.Embarked.apply(queenstown)
titanic_ted["Southamptom"] = titanic_ted.Embarked.apply(southamptom)
```

```python
titanic_ted.drop("Embarked",axis=1, inplace=True)
```

```python
titanic_ted["Cabin"] = titanic_ted.Cabin.apply(cabin_num)
```

```python
titanic_ted["Age"].fillna(titanic_ted["Age"].mean(), inplace = True)
```

```python
titanic_ted
```

```python
titanic_arr_ts = titanic_ted.values
```

```python
X_test_scaled = scaler.transform(titanic_arr_ts)
X_test_scaled.shape
```

```python
X_test_scaled = np.append(X_test_scaled,np.ones(X_test_scaled.shape[0]).reshape(-1,1),axis=1)
```

```python
Y_pred_test = alg.predict(X_test_scaled)
```
