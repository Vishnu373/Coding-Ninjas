```python
import matplotlib.pyplot as plt
```

```python
# dummy points as 2 arrays for x and y axis
x = [1, 2, 3]
y = [2, 4, 6]

# just shows the points in the graph
plt.scatter(x, y)

# acts like a print function in case of graphs
plt.show()
```

```python
# dummy points as 2 arrays for x and y axis
x = [1, 2, 3]
y = [2, 4, 6]

# connected by a line
plt.plot(x, y)
plt.show()
```

```python
# Combination of both scatter() and plot()
x = [1, 2, 3]
y = [2, 4, 6]

plt.scatter(x, y)
plt.plot(x, y)

plt.show()
```

```python
# creating 2 different lines
x = [1, 2, 3]
y = [2, 4, 6]

x2 = [3, 4, 5]
y2 = [2, 4, 6]

plt.scatter(x, y)
plt.scatter(x2, y2)

plt.plot(x, y)
plt.plot(x2, y2)

plt.show()
```

```python
# plot x^3 function
import numpy as np

# x = np.array([1, 2, 3, 4])

# another way of creating the array
x1 = np.arange(0, 5, 0.1) # arange(start, end, step_value)
y = x1 ** 3

plt.plot(x1, y)

plt.show()
```

```python
# Plotting graph with single axis
a = [3, 4, 5] # -> considers it as y points
# for x - points it takes range from 0 to number of points - 1
plt.plot(a)
plt.show()
```
