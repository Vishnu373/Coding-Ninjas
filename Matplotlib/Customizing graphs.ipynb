```python
import matplotlib.pyplot as plt
```

```python
# changing the color
x = [1, 2, 3]
y = [3, 4, 5]

# plt.plot(x, y, "b")
# plt.plot(x, y, "r")

plt.plot(x, y, color = "black")

plt.show()
```

```python
# changing the pattern
x = [1, 2, 3]
y = [3, 4, 5]

# plt.plot(x, y, "bo")
# plt.plot(x, y, "r--")
# plt.plot(x, y, "r+")
# plt.plot(x, y, "r^")


plt.plot(x, y, color = "black", marker = "+")

plt.show()
```

```python
# changing line width
x = [1, 2, 3]
y = [3, 4, 5]

plt.plot(x, y, color = "black", linewidth = 0.5)
# plt.plot(x, y, color = "black", linewidth = 5)

plt.show()
```

```python
# adding x and y label
x = [1, 2, 3]
y = [3, 4, 5]

plt.xlabel("x")
plt.ylabel("y")

plt.plot(x, y, color = "black")

plt.show()
```

```python
# adding title 
x = [1, 2, 3]
y = [3, 4, 5]

plt.title("Demo")

plt.plot(x, y, color = "black")

plt.show()
```

```python
# adding names to lines and legends
a = [1, 2, 3]
a_sqre = [1, 4, 9]
a_cube = [1, 8, 27]

plt.plot(a, a_sqre, color = "black", label = "a^2")
plt.plot(a, a_cube, color = "red", label = "a^3")

plt.legend()

plt.plot()
```

```python
# changing the axis values
x = [1, 2, 3]
y = [3, 4, 5]

plt.axis([0, 5, 3, 5])

plt.plot(x, y, color = "black")

plt.show()
```

```python
# For better understanding
x = [1, 2, 3]
y = [3, 4, 5]

plt.plot(x, y, color = "black")
plt.grid()

plt.show()
```

```python
# Adding some details at a particular point
x = [1, 2, 3]
y = [3, 4, 5]

plt.plot(x, y, color = "black")

plt.text(1, 3, "Lowest value", fontsize = 10)
plt.text(2, 4, "Mid value", fontsize = 10)
plt.text(3, 5, "Highest value", fontsize = 10)

plt.show()
```
