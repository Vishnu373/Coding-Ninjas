```python
import matplotlib.pyplot as plt
```

```python
# plotting pie graph
sizes = [3, 4, 6, 2] # an array of a number of splits

plt.pie(sizes)

plt.axis("equal") # makes it a perfect circle (optional)

plt.show()
```

```python
# changing the colors
# plotting pie graph
sizes = [3, 4, 6, 2] # an array of a number of splits
colors = ["blue", "red", "black", "yellow"] # this array  must be equal to the sizes array

plt.pie(sizes, colors = colors)

plt.show()
```

```python
# adding labels
sizes = [3, 4, 6, 2] # an array of a number of splits
labels = ["A", "B", "C", "D"] # this array  must be equal to the sizes array

plt.pie(sizes, labels = labels)

plt.axis("equal") # makes it a perfect circle (optional)

plt.show()
```

```python
# adding description to the graphs
sizes = [3, 4, 6, 2] # an array of a number of splits
labels = ["A", "B", "C", "D"] # this array  must be equal to the sizes array

plt.pie(sizes, labels = labels, autopct = "%.1f%%")

plt.axis("equal") # makes it a perfect circle (optional)

plt.show()
```

```python
# highlight a certain part
sizes = [3, 4, 6, 2] # an array of a number of splits
explode = [0.1, 0.2, 0, 0.5]

plt.pie(sizes, explode = explode, autopct = "%.1f%%", counterclock = False, startangle = 100)

plt.axis("equal") # makes it a perfect circle (optional)

plt.show()
```
